```python
alphabets = [chr(ord('a') + i) for i in range(26)]
num2char = {}
for i in range(10):
    if i < 2:
        num2char[str(i)] = [str(i)]
    elif i == 7 or i == 9:
        num2char[str(i)] = [alphabets.pop(0),alphabets.pop(0),alphabets.pop(0),alphabets.pop(0)]
    else:
        num2char[str(i)] = [alphabets.pop(0),alphabets.pop(0),alphabets.pop(0)]
        
```

```python
def number2words(number, output):
    if not number:
        print(output)
    else:
        last = number.pop()
        for x in num2char[last]:
            cnumber = number.copy()
            number2words(cnumber, x + output)
```

```python
NUMBER = "7996267"
number = list(NUMBER)
output = ''
```

```python
number2words(number, output)
```
