```python
class SymbolTable():
    def __init__(self):
        self.array =[]
        
        
    def put(self, key, value):
        self.array.append([key, value])
        sorted(self.array, key = lambda x: x[0])
    
    def binary_search(self, i, j, key):
        index = (i+j) // 2

        if i == index:
            if key == (self.array[i])[0]:
                return index
            return None
            
        if (self.array[index])[0] == key:
            return index
        elif (self.array[index])[0] > key:
            return self.binary_search(i, index - 1, key)
        else:
            return self.binary_search(index + 1, j, key)
    
    def get(self, key):
        return self.binary_search(0, len(self.array)-1, key)
    
    def delete(self, key):
        index = self.get(key)
        if(index == None):
            return False
        del self.array[index]
        return True
    
    def contains(self, key):
        if self.is_empty():
            print("isempty")
            return False
        if self.binary_search(0, len(self.array) - 1, key):
            return True
        return False
    
    def is_empty(self):
        return self.size() == 0
    
    def size(self):
        return len(self.array)
    
    def keys(self):
        return [item[0] for item in self.array]
    
    def maximum(self):
        return self.array[len(self.array) -1][1]
    
    def minimum(self):
        return self.array[0][1]
        
```

```python
class ST_BST():
    
    class Node():
        def __init__(self, key, value):
            self.right = None
            self.left = None
            self.key = key
            self.value = value
            self.size = 1
        def __str__(self):
            return (f"""key: {self.key} value: {self.value} size: {self.size} """
                    f"""left: {self.left} right: {self.right}""")
    def __init__(self):
        self.root = None
        
    def _put(self, root, key, value):
        if root == None:
            root = self.Node(key, value)
            return root
        else:
            if root.key == key:
                root.value = value
                return root
            elif key > root.key:
                found = self._put(root.right, key, value)
                root.right = found
            else:
                found = self._put(root.left, key, value)
                root.left = found
            return root
        
    def put(self, key, value):
        self.root = self._put(self.root, key, value)
        #print(self.root)
        
    def _get(self, root, key):
        if root == None:
            return root
        if root.key == key:
            return root
        elif key > root.key:
            return self._get(root.right, key)
        else:
            return self._get(root.left, key)
        
    def get(self, key):
        found = self._get(self.root, key)
        if found:
            return found.value
        return None
    
    def _delete(self, root, key):
        print(f"_delete called with {root.key}")
        
    def delete(self, key):
        self._delete(self.root, key)
        
    def contains(self, key):
        found = self._get(self.root, key)
        if found:
            return True
        return False
    
    def is_empty(self):
        return self.root == None
    
    def size(self):
        if not self.root:
            return 0
        return self.root.size
    
    def _keys(self, root):
        if not root:
            return []
        return self._keys(root.left) + [root.key] + self._keys(root.right)
    
    def keys(self):
        return self._keys(self.root)

        
    def ceiling_(self, root, val, key):
        if not root:
            return val
        if root.key == key:
            return key
        if root.key < key:
            return self.ceiling_(root.right,val ,key)
        if root.key > key:
            return self.ceiling_(root.left,root.value ,key)
        
        
    def ceil(self, key):
        return self.ceiling_(self.root, None, key)
    
    def floor_(self, root, val, key):
        if not root:
            return val
        if root.key == key:
            return key
        if root.key < key:
            return self.floor_(root.right,root.value ,key)
        if root.key > key:
            return self.floor_(root.left,val ,key)
        
        
        
    def floor(self, key):
        return self.floor_(self.root, None, key)
```
