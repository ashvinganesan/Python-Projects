```python
def get_intersection(line1, line2):
    # line1 = [[0, 0], [8, 0]]
    # line2 = [[0, 0], [0, 7]]
    lines = get_xline_and_yline(line1, line2)
    if not lines:
        return None
    if not is_xline_within_yline(lines):
        return None
    if not is_yline_within_xline(lines):
        return None
    return getpoint(lines)

def get_xline_and_yline(line1, line2):
    #if (line1[0][1] == line1[1][1]) and line2[0][0] == line2[1][0]:
    #return [line1, line2]
    #if (line2[0][1] == line2[1][1]) and line1[0][0] == line1[1][0]:
    #return [line2, line1]

    if (line1[0][0] == line1[1][0]) and line2[0][1] == line2[1][1]:
        return [line2, line1]
    if (line2[0][0] == line2[1][0]) and line1[0][1] == line1[1][1]:
        return [line1, line2]
    return None

def is_yline_within_xline(lines):
    # lines = [[[0, 0], [8, 0]], [[0, 0], [0, 7]]]
    xline = lines[0] # xline = [[3, 2], [5, 2]]
    yline = lines[1] # yline = [[1, 1], [1, 7]]
    if (min(xline[0][0], xline[1][0]) <= yline[0][0]) and (yline[0][0] <= max(xline[0][0], xline[1][0])):
        return True
    return False

def is_xline_within_yline(lines):
    # lines = [[[3, 2], [5, 2]], [[1, 1], [1, 7]]]
    xline = lines[0] # xline = [[3, 2], [5, 2]]
    yline = lines[1] # yline = [[1, 1], [1, 7]]
    if (min(yline[0][1], yline[1][1]) <= xline[0][1] and (xline[0][1] <= max(yline[0][1], yline[1][1]))):
        return True
    return False

def getpoint(lines):
    xline = lines[0] # xline = [[0, 0], [8, 0]]
    yline = lines[1]
    return [yline[0][0], xline[0][1]]


def parse_wire_path(unparsed):
    return [[x[0],int(x[1:])] for x in unparsed.split(",")]

def build_lines(path):
    lines = []
    x = 0
    y = 0
    for element in path: 
        if element[0] == "R":
            lines.append([[x,y],[x + element[1],y]])
            x += element[1]             
        if element[0] == "L":
            lines.append([[x,y],[x - element[1],y]])
            x -= element[1]
        if element[0] == "U":
            lines.append([[x,y],[x ,y + element[1]]])
            y += element[1]
        if element[0] == "D":
            lines.append([[x,y],[x ,y - element[1]]])
            y -= element[1]
    return lines
```
